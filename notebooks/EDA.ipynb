Analisis de datos   


```python
import pandas as pd
import plotly.graph_objects as go # Importación de plotly.graph_objects como go


# Leer los datos del archivo CSV
car_data = pd.read_csv('/Users/DasHa/anaconda_projects/Proyecto-7/vehicles_us.csv')
print('Archivo cargado exitosamente')
```

```python
# Un vistazo rapido al df
print(car_data.head(12))
```

Limpiando los Datos

```python
#Buscando datos repetidos
print("El numero de valores duplicados es : ",car_data.duplicated().sum())

# buscamos valores mal escritos
car_data['model_year']=car_data['model_year'].fillna('0')
car_data['model_year']=car_data['model_year'].astype('int')

car_data['cylinders']=car_data['cylinders'].fillna('0')
car_data['cylinders']=car_data['cylinders'].astype('int')

car_data['odometer']=car_data['odometer'].fillna('0')
car_data['odometer']=car_data['odometer'].astype('int')



# Primero veamos como se ven los datos antes de limpiarlos
#print("Antes de la limpieza:")
#print(repr(car_data['paint_color'].iloc[0:5].tolist()))

# Ahora apliquemos la limpieza paso a paso
car_data['paint_color'] = car_data['paint_color'].fillna('Sin color')
car_data['paint_color'] = car_data['paint_color'].str.strip()

# Veamos los mismos ejemplos después de la limpieza
#print("\nDespués de la limpieza:")
#print(repr(car_data['paint_color'].iloc[0:5].tolist()))
# Veamos TODOS los valores únicos con repr() para detectar espacios ocultos
#print("Valores únicos con repr():")
#unique_colors = car_data['paint_color'].unique()
#for color in unique_colors:
    #print(repr(color))

car_data['is_4wd']=car_data['is_4wd'].fillna('No')
car_data['is_4wd']=car_data['is_4wd'].astype('str')
car_data['is_4wd']=car_data['is_4wd'].replace('1.0','Si')
#print(car_data['is_4wd'].unique()) 
```

Histograma

```python
# Crear un histograma utilizando plotly.graph_objects
# Se crea una figura vacía y luego se añade un rastro de histograma
fig = go.Figure(data=[go.Histogram(x=car_data['odometer'], nbinsx=1000)])

# Opcional: Puedes añadir un título al gráfico si lo deseas
fig.update_layout(title_text='Distribución del Odómetro')
fig.update_xaxes(range=[0, 350000])  # Limitando el eje x de 0 a 0.2M ya que la mayor parte de los datos stan en este rango
fig.update_yaxes(range=[0, 900])
fig.update_layout(title_text='Distribución del Odómetro (0-0.2m)')

# Mostrar el gráfico Plotly
fig.show()
```

Grafico de dispersión

```python
# Crear un scatter plot utilizando plotly.graph_objects
# Se crea una figura vacía y luego se añade un rastro de scatter
fig = go.Figure(data=[go.Scatter(x=car_data['odometer'], y=car_data['price'], mode='markers')])

# Opcional: Puedes añadir un título al gráfico si lo deseas
fig.update_layout(title_text='Relación entre Odómetro y Precio')

# Mostrar el gráfico Plotly
fig.show()
```
